```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = ['a', 'b', 'c', 'd', 'e', 'f', 'g']
y = [10, 8, 12, 13, 15.5, 16, 17]

plt.figure(figsize=(5, 3), dpi=100)

plt.plot(x, y, linestyle='--', linewidth=2, color ='#011C40')

plt.title('Título', fontdict = {'fontsize':15})
plt.xlabel('Produto', fontsize=14)
plt.ylabel('Quantidade', fontsize=14)


plt.show()



```

```python
mes = ['Janeiro', 'Fevereiro',' Março', 'Abril', 'Março']
qtde = [200, 186, 158, 153, 179]
```

```python
plt.plot(mes, qtde, linestyle='--', linewidth=4, color ='#003185', marker='o', mfc='#00EBFF', markersize=8)

plt.xlabel('Mês')
plt.ylabel('Quantidade')
plt.show()
```

```python
x = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l']
y = [13, 12, 13, 10, 13.5, 9, 17, 18, 20, 22, 18, 23]

plt.figure(figsize=(7, 4), dpi=100)

plt.plot(x, y, linestyle='-', color ='#BF0B3B', linewidth=3)

plt.title('Título', fontdict = {'fontsize':20})
plt.xlabel('eixo x', fontsize=16)
plt.ylabel('eixo y', fontsize=16)

plt.tick_params(length=3, width=2)
plt.xticks(['a', 'b', 'e', 'g'])
plt.yticks([9, 12, 13, 17])



plt.show()
```
